```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import cv2
from skimage import color
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
dataDir='..'
dataType='val2017'
annFile='{}/annotations/instances_{}.json'.format(dataDir,dataType)
```

```python
# initialize COCO api for instance annotations
coco=COCO(annFile)
```

```python
# display COCO categories and supercategories
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories: \n{}\n'.format(' '.join(nms)))

nms = set([cat['supercategory'] for cat in cats])
print('COCO supercategories: \n{}'.format(' '.join(nms)))
```

```python
# get all images containing given categories, select one at random
catIds = coco.getCatIds(catNms=['person','dog','skateboard']);
imgIds = coco.getImgIds(catIds=catIds );
imgIds = coco.getImgIds(imgIds = [324158])
img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]
```

```python
# load and display image
# I = io.imread('%s/images/%s/%s'%(dataDir,dataType,img['file_name']))
# use url to load image
I = io.imread(img['coco_url'])
plt.axis('off')
plt.imshow(I)
plt.show()
```

```python
# load and display instance annotations
plt.imshow(I); plt.axis('off')
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
```

```python
def display_with_ann(categories, draw_bbox, b_and_w):
    # get all images containing given categories, select one at random
    catIds = coco.getCatIds(catNms=categories);
    imgIds = coco.getImgIds(catIds=catIds);
    # if no images with given categories then print and end
    if not len(imgIds):
        print('No images with all given categories')
        return
    img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]
    I = io.imread(img['coco_url'])

    # load and display instance annotations
    plt.imshow(I); plt.axis('off')
    annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
    anns = coco.loadAnns(annIds)
    coco.showAnns(anns, draw_bbox)
    # change title if bounding boxes are used
    if draw_bbox:
        plt.title('Image with Annotations and Bounding Boxes')
    else:
        plt.title('Image with Annotations')
    plt.show()
    
    #show isolated things
    #collect all the annotations, covert to an rgb image of 0s and 1s
    output_image = np.zeros((len(I), len(I[0]), 3))
    for ann in anns:
        grey_ann = coco.annToMask(ann)
        rgb_ann = cv2.cvtColor(grey_ann, cv2.COLOR_GRAY2RGB)
        scaled_ann = (rgb_ann > 0) * 1
        output_image += scaled_ann
    #make sure no pixels are over 1
    output_image[output_image > 1] = 1
    #element-wise multiplication to extract only annotated figures
    if b_and_w:
        output_image = ((output_image*I) > 0) * 255
        img = color.rgb2gray(output_image)
        plt.imshow((img).astype(int), cmap='gray')
    else:
        plt.imshow((output_image*I).astype(int))
    plt.axis('off')
    plt.title('Image without Backgroud')
```

```python
display_with_ann(['dog', 'sink'], draw_bbox=False, b_and_w=True)
```
